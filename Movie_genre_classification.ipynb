```python
import pandas as pd

# Corrected data loading logic (temporary for this cell's execution)
def _load_data_corrected(file_path):
  with open(file=file_path, mode="r", encoding="utf-8") as f:
    lines = f.readlines() # Read lines instead of entire content
  processed_data = [line.strip().split(' ::: ') for line in lines if line.strip()] # Process each line
  return processed_data

# Re-load the data correctly using the temporary corrected function
train_data = _load_data_corrected("/content/train_data.txt")
test_data = _load_data_corrected("/content/test_data.txt")
test_solution = _load_data_corrected("/content/test_data_solution.txt")

# Now create DataFrames with the correctly loaded data
train_df=pd.DataFrame(train_data,columns=['ID','Title','Genre',"Description"])
test_df=pd.DataFrame(test_data,columns=["ID","Title","Description"])
test_solution_df=pd.DataFrame(test_solution, columns=["ID","Title","Genre","Description"])
```

TF-**IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(max_features=1000000)

x_train_tfidf=vectorizer.fit_transform(train_df["Description"])
x_test_tfidf=vectorizer.transform(test_df["Description"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
y_train=label_encoder.fit_transform(train_df["Genre"])
label_encoder.classes_
```

# **Train** **Model**

```python
from sklearn.linear_model import LogisticRegression
lr_model= LogisticRegression(max_iter=2000)
lr_model.fit(x_train_tfidf,y_train)
```

# Predict

```python
y_pred=lr_model.predict(x_test_tfidf)
predicted_genres=label_encoder.inverse_transform(y_pred)
```

```python
test_df['Predicted_genres']=predicted_genres
test_df[['Title','Predicted_genres']]
```

```python
merged_df=pd.merge(test_solution_df[["ID","Genre"]],test_df[["ID","Predicted_genres"]],on="ID")
merged_df
```

# Evaluation

```python
from sklearn.metrics import accuracy_score , classification_report
accuracy=accuracy_score(merged_df["Genre"],merged_df["Predicted_genres"])
print(f"Accuracy:{accuracy}")

print(classification_report(merged_df["Genre"],merged_df["Predicted_genres"]))
```

# Model Building

```python
from sklearn.naive_bayes import MultinomialNB
nb_model=MultinomialNB()
nb_model.fit(x_train_tfidf,y_train)
```

```python
y_pred_nb=nb_model.predict(x_test_tfidf)
predicted_genres_nb=label_encoder.inverse_transform(y_pred_nb)
test_df['Predicted_genres_NB']=predicted_genres_nb
merged_df_nb=pd.merge(test_solution_df[["ID","Genre"]],test_df[["ID","Predicted_genres_NB"]],on="ID")
#
```

# Model Evaluation

```python
accuracy_nb=accuracy_score(merged_df_nb["Genre"],merged_df_nb['Predicted_genres_NB'])

print(f"Accuracy Naive Bayes:{accuracy_nb}")

print(classification_report(merged_df["Genre"],merged_df_nb["Predicted_genres_NB"],target_names=label_encoder.classes_))
```
